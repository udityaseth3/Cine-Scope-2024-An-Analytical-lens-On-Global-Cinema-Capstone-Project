INSIGHT

```python
# Top 10 Highest Grossing Movies
df.sort_values(by='Box Office ($M)', ascending=False)[['Movie Title', 'Box Office ($M)', 'Genre', 'Country']].head(10)
```

```python
# Genres with Highest Average IMDB Rating
df.groupby('Genre')['IMDB Rating'].mean().sort_values(ascending=False).head(10)
```

```python
# Countries Producing Most Movie In 2024
df['Country'].value_counts().head(10)
```

```python
# Languages with Highest Rated Movies
df.groupby('Language')['IMDB Rating'].mean().sort_values(ascending=False).head(10)
```

```python
# Correlation Matrix
df[['Budget ($M)', 'Box Office ($M)', 'IMDB Rating', 'Rotten Tomatoes (%)']].corr()
```

```python
# Platform-wise Movie Count
df['Platform'].value_counts()
```

```python
# Month-wise Movie Releases
df['Release Date'] = pd.to_datetime(df['Release Date'])
df['Month'] = df['Release Date'].dt.month
df['Month'].value_counts().sort_index()
```

```python
#  Unique Values in Each Column
for col in df.columns:
    print(f"\nUnique values in {col}:\n", df[col].unique())
```

Univariate Analysis

```python

#  Distribution Plots
num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

Bivariate Analysis

```python
# Relationship Between Budget and Box Office
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x='Budget ($M)', y='Box Office ($M)', hue='Genre')
plt.title("Budget vs Box Office Collection")
plt.show()
```

```python
# Correlation Heatmap (compatible with all versions)
numeric_df = df.select_dtypes(include=['number'])  # filter only numeric columns

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

BOX PLOT TO CHECK OUTLIERS IN DATA SET

```python

#  Boxplot for outlier detection
sns.boxplot(data=df, x='Genre', y='IMDB Rating')  # change as needed
plt.title("Boxplot of  Rating by Genre")
plt.show()

```

PAIR PLOT

```python
# Choose a subset of numeric columns
sns.pairplot(df)
plt.show()
```
